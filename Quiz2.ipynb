```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Quiz2.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

## **Clean the Dataset**

## 1-Correct Wrong Formats

```python

df['col2'] = pd.to_datetime(df['col2'], errors='coerce')
```

```python
df['col3'] = pd.to_numeric(df['col3'], errors='coerce')
```

```python
df
```

## 2-Handle Empty Cells

```python
df['col2'] = df['col2'].fillna(df['col2'].mean())
df['col3'] = df['col3'].fillna(df['col3'].mean()) 
df['col5'] = df['col5'].fillna(df['col5'].mean())
```

```python
df
```

## 3-Remove Duplicates

```python

df = df.drop_duplicates()
```

## **Data Visualization** 

```python
# Scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(df['col1'], df['col5'])
plt.title('Scatter Plot of col1 vs col5')
plt.xlabel('col1')
plt.ylabel('col5')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
df['col1'].value_counts().plot(kind='bar')
plt.title('Bar Plot of col1 Value Counts')
plt.xlabel('col1')
plt.ylabel('Counts')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
df['col5'].hist(bins=15)
plt.title('Histogram of col5')
plt.xlabel('col5')
plt.ylabel('Frequency')
plt.show()
```
